# Графики plotly

```python
import numpy as np
```

```python
np.arange(1, 10, 1.7)
```

```python
x = np.linspace(-10, 10, 1000)
y = np.sin(x)

fig = go.Figure() #создали фигуру, на которую собираемся навешивать графики
fig.add_trace(go.Scatter(x=x, y=y, mode='lines')) #добавили синусоиду
fig.show() #показали график
```

```python
def plot_sin(x_axis_data):
    y = np.sin(x_axis_data)
    fig = go.Figure() #эта переменная fig существует только внутри функции
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines'))
    return fig

fig = plot_sin(np.linspace(-10, 10, 1000)) # эта fig не имеет никакого отношения к предыдущей
fig.show()

```

```python
def plot_sin_cos(x_axis_data):
    # переделайте функцию выше так, чтобы она возвращала фигуру с графиками sin(x) и cos(x)
    y1 = np.sin(x_axis_data)
    y2 = np.cos(x_axis_data)
    fig = go.Figure() #эта переменная fig существует только внутри функции
    fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name = 'sinx' ))
    fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name = 'cosx'))
    return fig
    
fig = plot_sin_cos(np.linspace(-10, 10, 1000)) # эта fig не имеет никакого отношения к предыдущей
fig.show()
```

## Задание 5 (2 балла) 

```python
from plotly.subplots import make_subplots

def plot_sin_cos_subplots(x_axis_data):
    # переделайте функцию выше так, чтобы она возвращала фигуру с графиками sin(x) и cos(x) на отдельных подграфиках
    y1 = np.sin(x_axis_data)
    y2 = np.cos(x_axis_data)
    fig = make_subplots(rows=1, cols=2,
          subplot_titles =("sinx", "cosx")
           )
    fig.add_trace(
        go.Scatter(x=x, y=y1, mode='lines'),
        row=1, col=1
        )
    fig.add_trace(
        go.Scatter(x=x, y=y2, mode='lines'),
        row=1, col=2
        )
    fig.update_layout(
        showlegend =False
    )
    return fig

fig = plot_sin_cos_subplots(np.linspace(-10, 10, 1000)) # эта fig не имеет никакого отношения к предыдущей
fig.show()
```
