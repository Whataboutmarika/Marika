```python
import numpy as np

n = [5, 10, 20, 30, 50]

A1 = np.array([[1/4, 3/4, 0, 0], [1/2, 0, 1/2, 0], [0, 1/2, 0, 1/2], [0, 0, 3/4, 1/4]])

A2 = np.array([[2/9, 1/3, 0, 4/9], [4/9, 1/9, 0, 4/9], [2/9, 2/9, 2/9, 1/3], [2/9, 2/9, 4/9,1/9]])

A3 = np.array([[4/9, 0, 2/9, 1/3], [0, 4/9, 2/9, 1/3], [0, 2/3, 0, 1/3], [4/9, 2/9, 2/9, 1/9]])

def stationary_distribution(A, k):
    n = np.shape(A)[1]
    v0 = np.zeros(n-1)
    v0 = np.append(v0,1)
    v1 = np.ones(n-1)
    v1 = np.append(v1,2)
    A1 = np.delete(A2, n-1, axis = 0)
    M1 = np.vstack((A1, v1)) - np.eye(n)
    solve = np.linalg.solve(M1, v0)
    print("Стационарное распределние матрицы A", k,':', solve)

def power(A, n, k):
    for i in n:
        l = np.linalg.matrix_power(A, i)
        print('Матрица A',k ,'в степени', i, ':')
        print(l, '\n')


print(stationary_distribution(A1,1),'\n')
print(stationary_distribution(A2, 2),'\n')
print(stationary_distribution(A3, 3),'\n')


print(power(A1, n, 1))
print(power(A2, n, 2))
print(power(A3, n, 3))
```
