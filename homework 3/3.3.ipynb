```python
import  numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


repetition = 1000

n1 = 100
p1 = 0.04

lambda1 = 4

bin = np.random.binomial(n1, p1, size = repetition)
pois = np.random.poisson(lam = lambda1, size = repetition)


fig = plt.figure()
ax = fig.add_subplot(111)
fig.set(facecolor = 'lightcyan')
ax.set(xlim = [0, 12],
       ylim = [0, 300],
       title = 'Биномиальное распределение')
plt.hist(bin,color = 'm')
plt.show()


fig = plt.figure()
ax = fig.add_subplot(111)
fig.set(facecolor = 'lightcyan')
ax.set(xlim = [0, 12],
       ylim = [0, 300],
       title = 'Распределение Пуассона')
plt.hist(pois,color = 'blueviolet')
plt.show()


fig = plt.figure()
ax = fig.add_subplot(111)
fig.set(facecolor = 'lightcyan')
ax.set(xlim = [0, 12],
       ylim = [0, 300],
       title = '2 распределения')
plt.hist(bin,color = 'm')
plt.hist(pois,color = 'blueviolet')
plt.show()

```

```python

n1 = 100
p1 = 0.04

n2 = 1000
p2 = 0.004

bin1 = np.random.binomial(n1, p1, size = repetition)
bin2 = np.random.binomial(n2, p2, size = repetition)


fig = plt.figure()
ax = fig.add_subplot(111)
fig.set(facecolor = 'lightcyan')
ax.set(xlim = [0, 12],
       ylim = [0, 300],
       title = 'Биномиальное распределение(100, 0.04)')
plt.hist(bin1,color = 'm')
plt.show()


fig = plt.figure()
ax = fig.add_subplot(111)
fig.set(facecolor = 'lightcyan')
ax.set(xlim = [0, 12],
       ylim = [0, 300],
       title = 'Биномиальное распределение(1000, 0.004)')
plt.hist(bin2,color = 'blueviolet')
plt.show()


fig = plt.figure()
ax = fig.add_subplot(111)
fig.set(facecolor = 'lightcyan')
ax.set(xlim = [0, 12],
       ylim = [0, 300],
       title = '2 распределения')
plt.hist(bin1,color = 'm')
plt.hist(bin2,color = 'blueviolet')
plt.show()
```

```python
repetition = 1000

n1 = 100
p1 = 0.04

bin = np.random.binomial(n1, p1, size = repetition)

bin_norm = (bin - n1*p1)/(n1*p1*(1 - p1))**(1/2)

fig = plt.figure()
ax = fig.add_subplot(111)
fig.set(facecolor = 'lightcyan')
x = pd.Series(bin_norm, name="x variable")
sns.distplot(x)
plt.show()
```
